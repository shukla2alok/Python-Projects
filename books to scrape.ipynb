## Load the HTML 

```python
# Importing libraries
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get('https://books.toscrape.com/')
response
```

```python
response.headers
```

```python
html_data = response.text
```

```python
html_data
```

## Parse the HTML

```python
data = BeautifulSoup(html_data, 'html.parser')
```

```python
data
```

```python
print(data.prettify())
```

## Locate and extract the data

```python
# object.tag_name
data.title
```

```python
data.title.string
```

```python
# getting first hyperlink -> object.a
data.a
```

```python
data.a.string
```

```python
# searching with the help of class
b1 = data.find(class_ = 'product_pod')
```

```python
b1
```

```python
# extracting the title
b1.h3
```

```python
b1.h3.a
```

```python
# title is a attribute of a tag
b1.h3.a['title']
```

```python
# href is also a attribute of a tag
b1.h3.a['href']
```

```python
# There are so many hyperlinks available in the website we are going to fetch only hyperlink for books
# we are going to do this by using the class of the books
```

```python
books = data.find_all(class_ = 'product_pod')
```

```python
books
```

```python
print(len(books))
```

```python
# There are 20 books inside a web page
# we go step wise to get the hyperlinks of all books
```

```python
for i in books:
    print(i.h3)
```

```python
for i in books:
    print(i.h3.a)
```

```python
for i in books:
    print(i.h3.a['href'])
```

```python
# These are all relative hyperlinks we want to print the correct hyperlink
```

```python
for i in books:
    print('https://books.toscrape.com/' + i.h3.a['href'])
```

```python
# doing it in more structured way
base_url = 'https://books.toscrape.com/'
books_url = []
for i in books:
    books_url.append(base_url + i.h3.a['href'])
```
